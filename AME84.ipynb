```python
def UnitaryMatrix(U,N):
    return U.ConjugateTranspose(U)-np.eye(N)
```

```python
def ConjugateTranspose(self):
    return self.conj().T
```

```python
def UnitaryMatrixQ(U,N):
    return np.allclose(np.dot(U,ConjugateTranspose(U)),np.eye(N),rtol=0,atol=1e-06)
```

```python
def Upol(U):
    return np.dot(np.linalg.svd(U)[0],np.linalg.svd(U)[2])
```

```python
def EntVec(U):
    l={1,2,3,4,5,6,7}
    BiP=list(itertools.combinations({1,2,3,4,5,6,7},3))
    EntU=[]
    for x in BiP :
        indexlist=(0,)+x+tuple(l.difference(set(x)))
        EntU.append(Ent(U,indexlist))
    return EntU                
```

```python
def Ent(U,indexlist):
    d=4
    UR=Reshape(U,indexlist)
    Unew=la.matrix_power(np.dot(UR,ConjugateTranspose(UR)),2)
    return 1/(1-1/d**4)*(1-1/d**8.0*Unew.trace())
```

```python
d=4
count=0
for per in permutations([0,1,2,3,4,5,6,7]):
    UR=AME416.reshape([d,d,d,d,d,d,d,d]).transpose(per).reshape(N,N)
    Unew=la.matrix_power(np.dot(UR,ConjugateTranspose(UR)),2)
    En=1/(1-1/d**4)*(1-1/d**8.0*Unew.trace())
    if(En>=0.9999999999):
        count+=1
print(count)    
```
